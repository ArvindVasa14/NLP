## Stemming

```python
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
# lower casing
final_para= final_para.lower()
```

```python
# tokenization
sentences= sent_tokenize(final_para)

tokenized_sent= []
for sent in sentences:
    tokenized_sent.append(' '.join([word for word in word_tokenize(sent)]))
tokenized_sent
```

```python
# stemming

stemming= PorterStemmer()
stemmed_sent= []
for sent in removed_special_sent:
    stemmed_sent.append(' '.join([stemming.stem(word) for word in word_tokenize(sent)]))
stemmed_sent
```

```python
# lammatization

from nltk.stem import WordNetLemmatizer
lemm= WordNetLemmatizer()

lammatized_sent= []
for sent in removed_special_sent:
    lammatized_sent.append(' '.join([lemm.lemmatize(word) for word in word_tokenize(sent)]))
lammatized_sent
```
