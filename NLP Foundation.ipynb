```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
example= """By contrast, Amazon last autumn said it was building out a system with 100,000 of its Trainium2 chips, while Oracle brought online a system with 32,000 Nvidia H100 GPUs.

A Meta spokesperson declined to specify which GPU suppliers the company was using aside from market leader Nvidia, though it has said publicly it also plans to use chips from AMD. Reuters has previously reported it has an internally designed GPU-like chip in the works, as well.

Meta has scrambled to build out its computing arsenal to support a mobilisation around generative AI this past year, after years of producing leading research on the technology through its FAIR team but scant focus on building it into its core social media products and AR/VR hardware devices.

The company convened a "GenAI" team last year to lead an effort to change that after the breakout success of OpenAI's ChatGPT chatbot in late 2022.

Since then, Meta has launched a commercial version of its Llama large language model, ad tools that can generate image backgrounds from text prompts and a "Meta AI" chatbot that can be accessed directly via its Ray-Ban smart glasses.

In his posts on Thursday, Zuckerberg said the company was currently training a third version of the Llama model.

He also tied the AI investments back to the pivot to the AR/VR-driven metaverse vision that inspired him to change the company's name to Meta in 2021, saying people would "need new devices" like glasses for interacting with AI.

In an interview with The Verge outlet, Zuckerberg also said Meta was striving for general intelligence in AI so as to develop more products."""
```

```python
example
```

```python
sent_tokenize(example)
```

```python
for sent in sent_tokenize(example):
    print(sent)
```

```python
word_tokenize(example)
```

```python
for word in word_tokenize(example):
    print(word)
```

## White space tokenizer

```python
white_token= nltk.tokenize.WhitespaceTokenizer()
# removes all unwanted special chars like , '' "" 
```

```python
white_token.tokenize(example)
# works like word tokenizer ignoring special chars
```

```python
for white in white_token.tokenize(example):
    print(white)
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
word_stop= nltk.corpus.stopwords.words('english')
word_stop
```

```python
word_cloud= WordCloud(max_words=100, stopwords= word_stop)
word_cloud.generate(para2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,14))
plt.imshow(word_cloud)
```

```python
# manually including into stop words
word_cloud= WordCloud(max_words=100, stopwords= word_stop + ['take','also','get','could'])
word_cloud.generate(para2)
```

```python
plt.figure(figsize=(16,14))
plt.imshow(word_cloud)
```

```python
word_cloud= WordCloud(background_color='white',max_words=100, stopwords= word_stop + ['take','also','get','could'])
word_cloud.generate(para2)
plt.figure(figsize=(16,14))
plt.imshow(word_cloud)
```
